## 代码

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=list(range(1,8))
y= [50, 51, 52, 48, 47, 49, 46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
# dashdot
plt.plot(x,y, color = 'green', linewidth =4, linestyle ='dotted')
```

```python
plt.plot(x, y, 'g+-.')
```

```python
plt.plot(x,y,color = 'red', marker = 'D', linestyle='')
```

```python
plt.plot(x,y,color = '#008040', marker = '+', linestyle='', markersize =10)
```

```python
plt.plot(x,
         y,
         color = '#008040',
         marker = '+',
         alpha=0.5,
         markersize =10)
```

### Axes labels, Legend, Grid

```python
days = [1,2,3,4,5,6,7]
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(days, max_t, label= "Max")
plt.plot(days, min_t, label = "Min")
plt.plot(days, avg_t, label = "Avg")

# plt.legend(loc= "upper right")
plt.legend(loc = 'best', shadow=True, fontsize = 'large')
plt.grid()
```

### Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['GOOGLE', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
profit = [40, 2, 34, 12]
```

```python
# company is string, not good
plt.bar(company, revenue)
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos, revenue, label= 'Revenue')
plt.bar(xpos, profit, label = "Profit")
plt.legend()
```

```python
plt.xticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos - 0.2, revenue, width = 0.4,label= 'Revenue')
plt.bar(xpos + 0.2, profit, width = 0.4, label = "Profit")
plt.legend()
```

```python
plt.xticks(xpos, company)

plt.title("US Tech Stocks")
plt.barh(xpos - 0.2, revenue, label= 'Revenue')
plt.barh(xpos + 0.2, profit, label = "Profit")
plt.legend()
```

### Histograms

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.hist(blood_sugar, bins=3, rwidth = 0.95)
```

```python
plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth = 0.95, color ='g', histtype = 'step')
```

```python
plt.xlabel("sugar range")
plt.ylabel("Total number of patient")
plt.title('Blood Sugar analysis')

plt.hist([blood_sugar_men, blood_sugar_women],
         bins=[80, 100, 125, 150], 
         rwidth = 0.95, 
         color =['green', 'orange'], label =['Men', 'Women'])

plt.legend()
```

```python
plt.xlabel("sugar range")
plt.ylabel("Total number of patient")
plt.title('Blood Sugar analysis')

plt.hist([blood_sugar_men, blood_sugar_women],
         bins=[80, 100, 125, 150], 
         rwidth = 0.95, 
         color =['green', 'orange'],
         label =['Men', 'Women'], orientation='horizontal')

plt.legend()
```

### Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ['Home Rent', 'food', "Phone/Internet Bill", "Car", "Other Utilities"]
```

```python
plt.axis('equal')
plt.pie(exp_vals,
        labels= exp_labels,
        radius=1.5,
        autopct='%0.2f%%',
        shadow = True, explode = [0,0.2,0,0,0])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,
        labels= exp_labels,
        autopct='%0.2f%%',
        shadow = True,
        startangle= 45)
plt.show()
```

### Save Chart To a File Using savefig

```python
plt.axis('equal')
plt.pie(exp_vals,
        labels= exp_labels,
        autopct='%0.2f%%',
        shadow = True,
        startangle= 45)
plt.savefig('piechart.png',
            bbox_inches = 'tight',
            pad_inches = 2, transparent = True)
```
